# Les boucles indéfinies ou conditionnelles
L'instruction while

```python
while instruction : # ne pas oublier les :
    bloc d'instruction # Attenton à l'indentation
suite du programme
```

```python
nombre = 1
while nombre < 10:

    print("nombre de tour" , nombre)
    nombre = nombre + 1 # on peut écrire nombre +=1
print("on a fait ",nombre - 1, "tours")
```

```python
# Attention à ne pas créer de boucle sans fin.
# Par exemple : 
nombre = 5
while nombre != 100:
    nombre +=10
    print("nombre de tours ", nombre)
print("Fin")
```

step : indique l'incrémentation (optionnelle)

# Les listes

```python
ma_liste = []
print(type(ma_liste))
```

```python
ma_liste = ["Orange","Pomme","Tomate","Cerise"]
print("Afficher les éléments de la liste")
for n in ma_liste :
    print(n)
print("\n") 

print("Afficher la longueur de la liste")
print(len(ma_liste))
print("\n")

# On peut accèder directement à une valeur de la liste. Attention la liste commence à la position 0
print("Extraire le second élément de la liste")
print("La valeur en seconde position ",ma_liste[1])
print("\n")

# Ajouter un élément (à la fin de la liste)
print("Ajout d'un élément à la liste")
ma_liste.append('Poire')
print(ma_liste)
print("\n")

# Ajoute à la troisième place
ma_liste.insert(2,'Melon')
print("Ajout de Melon à la troisième place")
print(ma_liste)
print("\n")

# Trier la liste
print("Trie de la liste")
ma_liste.sort()
print(ma_liste)
```
